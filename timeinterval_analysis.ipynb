```python
import pandas as pd
from scipy.spatial import KDTree
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from datetime import datetime
import pickle
from matplotlib import cm
from math import sin, cos, sqrt, atan2, radians
import matplotlib.dates as mdates
import scipy.stats as stats



def crack(storm, start_time, end_time):
    file_path = f'Intraday_data/{storm}.csv'
    data = pd.read_csv(file_path)
    # Convert date and time to a datetime format in UTC [get day, then add to find market open in UTC, then subtract the time variance from then]
    data['datetime'] = pd.to_datetime(data['date']) + timedelta(hours = 13.5) + pd.to_timedelta(data['time'], unit='min')
    # Convert HO and RB prices to per barrel
    data.loc[data['sym'].str.contains('HO') | data['sym'].str.contains('RB'), ['bid', 'ask', 'prc']] *= 42

    contract = data.iloc[-1]['sym'][2:]
    ho = f'HO{contract}'
    rb = f'RB{contract}'
    cl = f'CL{contract}'

    
    if (storm == 'AL132020') | (storm == 'AL142020'):
        contract = 'V0'

    if storm == 'AL032017':
        contract = 'Q7'

    if storm == 'AL022019':
        contract = 'N9'

    if (storm == 'AL032020'):
        contract = 'N0'

    if (storm == 'AL082020'):
        contract = 'Q0'

    if (storm == 'AL192020'):
        contract = 'V0'

    if (storm == 'AL262020'):
        contract = 'X0'
        
    if (storm == 'AL282020'):
        contract = 'Z0'

    if (storm == 'AL162017'):
        contract = 'X7'

    if (storm == 'AL092012'):
        contract = 'V2'
        
    if (storm == 'AL042012'):
        contract = 'Q2'

    ho = f'HO{contract}'
    rb = f'RB{contract}'
    cl = f'CL{contract}'
    
    # Filter data for the specified time range and specific contracts
    filtered_data = data[(data['datetime'] >= start_time) & (data['datetime'] <= end_time)]
    filtered_data = filtered_data[filtered_data['sym'].isin([cl, rb, ho])]
        
    # Pivot the data for easier calculation
    pivoted_data = filtered_data.pivot_table(index='datetime', columns='sym', values='prc')

    # Calculate the 3:2:1 crack spreads
    pivoted_data['321crack_spread'] = -1 * (3 * pivoted_data[cl] - 2 * pivoted_data[rb] - pivoted_data[ho])

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(pivoted_data.index, pivoted_data['321crack_spread'], label=f'{cl}/{rb}/{ho} Crack Spread')
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator())
    plt.xticks(rotation=45)
    plt.xlabel('DateTime')
    plt.ylabel('3:2:1 Crack Spread')
    plt.title(f'3:2:1 Crack Spread for {contract} Contracts ({start_time.strftime("%B %d, %Y")} - {end_time.strftime("%B %d, %Y")})')
    plt.legend()
    plt.tight_layout()
    plt.show()
    print(pivoted_data['321crack_spread'])

    return pivoted_data

# Updated Data with Storm Names as ALDDDDDD
storm_data = [
    {"Storm Name": "AL092017", "Initial": "8/17", "End": "8/31", "Year": 2017},
    {"Storm Name": "AL132020", "Initial": "8/19", "End": "8/29", "Year": 2020},
    {"Storm Name": "AL192020", "Initial": "9/11", "End": "9/17", "Year": 2020},

    {"Storm Name": "AL022019", "Initial": "6/10", "End": "6/15", "Year": 2019},
    {"Storm Name": "AL162017", "Initial": "10/4", "End": "10/8", "Year": 2017},
    {"Storm Name": "AL092012", "Initial": "8/21", "End": "8/30", "Year": 2012},
    {"Storm Name": "AL032020", "Initial": "6/1", "End": "6/10", "Year": 2020},

    {"Storm Name": "AL142020", "Initial": "8/20", "End": "8/25", "Year": 2020},
    {"Storm Name": "AL032017", "Initial": "6/19", "End": "6/22", "Year": 2017},
    {"Storm Name": "AL142021", "Initial": "9/12", "End": "9/17", "Year": 2021},
    {"Storm Name": "AL222020", "Initial": "9/17", "End": "9/25", "Year": 2020},

    {"Storm Name": "AL082020", "Initial": "6/22", "End": "6/27", "Year": 2020},

    {"Storm Name": "AL262020", "Initial": "10/4", "End": "10/11", "Year": 2020},
    {"Storm Name": "AL282020", "Initial": "10/24", "End": "10/29", "Year": 2020},


    {"Storm Name": "AL042012", "Initial": "6/23", "End": "6/27", "Year": 2012}
]

# Correctly creating the dictionary
storm_dates_al = {
    storm["Storm Name"]: {
        "Initial": pd.to_datetime(f"{storm['Initial']}/{storm['Year']}", format="%m/%d/%Y"),
        "End": pd.to_datetime(f"{storm['End']}/{storm['Year']}", format="%m/%d/%Y")
    } for storm in storm_data
}

# Initialize 12 arrays for each 1-hour interval starting every 30 minutes
hour_intervals = [np.array([]) for _ in range(12)]

def squared_difference(time_point, df, k):
    if time_point in df.index and time_point + pd.Timedelta(minutes = 60) in df.index:
        start_value = df.loc[time_point, '321crack_spread']
        end_value = df.loc[time_point + pd.Timedelta(minutes = 60), '321crack_spread']
        return k* (end_value - start_value) ** 2
    else:
        return None

for storm in storm_dates_al:
    start_time = storm_dates_al[storm]["Initial"]
    end_time = storm_dates_al[storm]["End"]

    data = crack(storm, start_time, end_time)

    # Loop through the crack spread data at 30-minute intervals
    time_range = pd.date_range(start=data.index.min(), end=data.index.max() - pd.Timedelta(minutes = 60), freq='30T')
    for time_point in time_range:
        k = 1 
        interval_index = int(((time_point.hour * 60 + time_point.minute) // 30) % 12)
        diff = squared_difference(time_point, data, k)
        if ((diff is not None)):
            hour_intervals[interval_index] = np.append(hour_intervals[interval_index], diff)


# Calculating means
interval_means = [np.nanmean(interval) for interval in hour_intervals]

# Calculate variances for each interval
interval_variances = [np.nanvar(interval, ddof=1) for interval in hour_intervals]  # ddof=1 for sample variance
# Calculate standard deviation for each interval
interval_stds = [np.nanstd(interval, ddof=1) for interval in hour_intervals]

# Plotting
plt.figure(figsize=(10, 6))
midpoint_labels = ['30m', '60m', '90m', '120m', '150m', '180m', '210m', '240m', '270m', '300m', '330m', '0m']
plt.rcParams.update({'font.size': 12, 'font.family': 'Arial'})

# Calculating error bar limits
lower_error = [0] * len(interval_means)  # Lower error is zero
upper_error = [std for std in interval_stds]  # Upper error is the standard deviation

# Combining lower and upper error limits
asymmetric_error = [lower_error, upper_error]

plt.bar(midpoint_labels, interval_means, yerr=asymmetric_error, capsize=5, alpha=0.7, color='steelblue')


# # Plotting the bars for mean squared differences
# plt.bar(midpoint_labels, interval_means, yerr=interval_variances, capsize=5, alpha=0.7, label='Mean Squared Difference', color='steelblue')

# Overlaying individual data points using a scatter plot
for i, interval in enumerate(hour_intervals):
    x_values = np.random.normal(i, 0.04, size=len(interval))
    plt.scatter(x_values, interval, color='darkred', alpha=0.7, s=20)

plt.xlabel('Interval Midpoint')
plt.ylabel('Mean Squared Difference')
plt.gca().set_ylim(0, 0.7)
plt.title('Mean Squared Differences of Crack Spread Across Intervals for All Storms')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Saving the figure in high resolution
plt.savefig('plot.png', dpi=300)

plt.show()

# Number of intervals
num_intervals = len(hour_intervals)

# Initialize a matrix to store the p-values of the tests
p_values_matrix = np.zeros((num_intervals, num_intervals))

# Performing pairwise tests using the two-sample Kolmogorov-Smirnov test
for i in range(num_intervals):
    for j in range(i + 1, num_intervals):
        ks_stat, p_value = stats.ks_2samp(hour_intervals[i], hour_intervals[j])
        p_values_matrix[i, j] = p_value

# Print the matrix of p-values
print("Matrix of p-values for pairwise Kolmogorov-Smirnov tests:")
print(np.round(p_values_matrix, 3))

for i in range(num_intervals):
    print(np.nanmean(hour_intervals[i]))




```
